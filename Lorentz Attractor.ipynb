# Lorentz attractor numerical solution
Equations:

dx/dt = a(y - x)

dy/dt = x(b - z) - y

dz/dt = xy - cz

I tried to make this as simple as possible.

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib widget
```

```python
# Initial params
x0 = 1.0
y0 = 2.0
z0 = 1.0
a = 10.0
b = 28.0
c = 8.0 / 3.0

# Simulation parameters
dt = 0.01
t_sim = 50.0
t = np.linspace(0.0, t_sim, int(t_sim / dt))
```

```python
x, y, z = [], [], []
for time_point in t:
    x.append(x0)
    y.append(y0)
    z.append(z0)
    
    dx = dt * a * (y0 - x0)
    x0 = x0 + dx
    dy = dt * (x0 * (b - z0) - y0)
    y0 = y0 + dy
    dz = dt * (x0 * y0 - c * z0)
    z0 = z0 + dz
```

```python
ax = plt.axes(projection='3d')
max_t = int(t_sim / dt)
ax.scatter(x, y, z, c=t, cmap='copper', s=0.4)
plt.show()
```
